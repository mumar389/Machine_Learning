```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
```

```python
np.random.seed(232535)
```

```python
mean_vec1=np.array([0,0,0])
cov_mat1=np.array([[1,0,0],[0,1,0],[0,0,1]])
class1=np.random.multivariate_normal(mean_vec1,cov_mat1,100)
```

```python
mean_vec2=np.array([1,1,1])
cov_mat2=np.array([[1,0,0],[0,1,0],[0,0,1]])
class2=np.random.multivariate_normal(mean_vec1,cov_mat1,100)
```

```python
from mpl_toolkits.mplot3d import Axes3D,proj3d
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111,projection='3d')
ax.plot(class1[:,0],class1[:,1],class1[:,2],'o')
ax.plot(class2[:,0],class2[:,1],class2[:,2],'^')
plt.show()
```

**S-1:finding the cov matrix which rep cov b/w features**

```python
all_data=np.concatenate((class1,class2))
```

```python
all_data_t=all_data.T
#this cov b/w data but we want b/w features
cov=np.cov(all_data_t)
cov
```

**S-2:apply SVD**

```python
eign_val,eign_vec=np.linalg.eig(cov)
eign_val,eign_vec
```

```python
eign_val_vec_pair=[]
for i in range(len(eign_val)):
    eign_vect=eign_vec[:,i]
    eign_val_vec_pair.append((eign_val[i],eign_vect))
eign_val_vec_pair.sort(reverse=True)
eign_val_vec_pair
```
