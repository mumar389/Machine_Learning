Pandas Functions

```python
import pandas as pd
```

```python
iris=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
iris
```

```python
df=iris.copy()
df.shape
df.columns=["sl","sw","pl","pw","flower_type"]
```

```python
df.head(3)
```

Removing columns using pandas

```python
#removing rows
#drop(label,inplace) doesnot do it inplace, it makes a copy
# df.drop(0)
df.drop(0,inplace=True)
#the labels does not change when deleted
#when we dropped 0 th row then we can see 0 row is not in the dataframe
df.head()
```

```python
df.head(3)
```

```python
df.index
```

```python
#it gives us the 0 th label from the dataframe, which in our case is 1
df.index[0]
```

```python
#to remove by position follow below methd
# df.drop(df.index[0],inplace=True)
#dropping multiple rows at a time
df.drop(df.index[[0,1]],inplace=True)
```

```python
df.head()
```

```python
#it gives  what all rows greater than 3
ans=df.sl>3
ans.sum()#it gives count
df[df.sl>5]
```
