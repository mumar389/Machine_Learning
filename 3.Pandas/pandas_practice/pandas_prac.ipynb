**Pandas is different from numpy as it uses excel like approach to work with data whereas numpy uses array approach**

```python
import pandas as pd
import numpy as np
```

```python
iris=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
iris.shape
```

```python
df=iris.copy()
df.shape
df.columns=["sl","sw","pl","pw","flower_type"]
```

```python
# df.head()
# df.head(15)
df.iloc[:,:]
```

```python
#M-1 to find sum row-wise
#storing row-wise sum in a new column 'sum'
df['sum']=df.sl+df.sw+df.pl+df.pw
```

```python
df.head(10)
```

```python
#we cannot do normal row-wise sum using sum() because it has flower_type which is a string. so we need to replace it with a numeric type
def f(data):
    if(data=="Iris-setosa"):
        return 0
    elif(data=="Iris-versicolor"):
        return 1
    elif(data=="Iris-virginica"):
        return 2
    else:
        return 3
df['f_type']=df.flower_type.apply(f)
```

```python
del df['sum']
del df['flower_type']
df
```

```python
#M-2:
ans=[]
for i in range(df.shape[0]):
    ans.append(df.iloc[i,:].sum())
ans    
```
