```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[2,4,6]
#it picks random color each time
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
v=[3,4,5]
z=[2,4,6]
plt.scatter(x,y)
plt.plot(v,z,"g^")
plt.plot(x,y,"y")
```

```python
##plotting x^3
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
v=x**4
plt.plot(x,y,label="x^3")
plt.plot(x,v,label="x^2")
# plt.axis([0,10,0,800])
plt.grid()
plt.text(2,200,"Umar",fontsize=14)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Matplotlib Practice")
```

```python
a=[1,2,3]
plt.plot(a,color="green",marker=".")
##it is going to take x=[0,1,2] i. e range(n) by default
```

```python
#pie graph-:
sizes=[3,4,6,2]
labels=["Bike","Car","Scooty","Jeep"]
colors=["blue","red","purple","pink"]
explode=[0.2
         ,0,0,0]
plt.pie(sizes,colors=colors,labels=labels,autopct="%0.2f%%",explode=explode,counterclock=False,startangle=150)
plt.title("Car Data")
plt.axis("equal")
```
