```python
import numpy as np
def cost(x,y,m,c):
    row,col=x.shape
    cost_total=0
    for i in range(row):
        for j in range(col):
            cost_total+=(1/row)*(y[i]-(m[j]*x[i][j])+c)
    return cost_total     
def find_slope(x,y,m,j):
    m_slope=0
    row=x.shape[0]
    cols=x.shape[1]
    M=x.shape[0]

    for i in range(row):
        for k in range(cols):
            m_slope+=(-2/M)*(y[i]-m[k]*x[i][k])*x[i][j]
    return m_slope    
def step(x,y,lr,m):
    row=x.shape[0]
    cols=x.shape[1]
    m_new=np.zeros(cols)
    m_slope=0
    for j in range(cols):
        m_new[j]=m[j]-lr*find_slope(x,y,m,j)   
    return m_new   
def gd(x,y,lr,it,x_orig):
    l=x.shape[1]
    m=np.zeros(l)
    c=m[l-1]
    for i in range(it):
        m=step(x,y,lr,m)
        cost_n=cost(x_orig,y,m,c)
        # print("Cost at it-:",cost_n,it)
    return m    
def predict(x,m):
    n=x.shape[1]
    c=m[n]
    m_f = m[:13]
    c_prod=np.dot(x,m_f)
    return c_prod+c    
   

def run():
    data_test=np.loadtxt("test.csv",delimiter=",")
    x_test=data_test[:,:13]
    data=np.loadtxt("train.csv",delimiter=",")
    x=data[:,:13]
    y=data[:,13]
    rows, cols = x.shape
    ones_column = np.ones((rows, 1), dtype=x.dtype)
    X = np.concatenate((x, ones_column), axis=1)
    iterations=15
    learning_rate=0.001
    m=gd(X,y,learning_rate,iterations,x)
    y_pred=predict(x_test,m)
    # print("M",m)
    print(y_pred)
    # print(X)
    np.savetxt("predicted.csv",y_pred,delimiter=",",fmt='%.5f')
run()    
```
