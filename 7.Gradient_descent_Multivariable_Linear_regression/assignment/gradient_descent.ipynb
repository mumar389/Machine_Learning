```python
import numpy as np
def cost(x,y,m,c):
    row,col=x.shape
    cost_total=0
    for i in range(row):
        for j in range(col):
            cost_total+=(1/row)*(y[i]-(m[j]*x[i][j])+c)
    return cost_total  
def find_slope(x,y,m,j):
    m_slope=0
    row=x.shape[0]
    cols=x.shape[1]
    M=x.shape[0]

    for i in range(row):
        for k in range(cols):
            m_slope+=(-2/M)*(y[i]-m[k]*x[i][k])*x[i][j]
    return m_slope    
def step(x,y,lr,m):
    row=x.shape[0]
    cols=x.shape[1]
    m_new=np.zeros(cols)
    m_slope=0
    for j in range(cols):
        m_new[j]=m[j]-lr*find_slope(x,y,m,j)   
    return m_new   
def gd(x,y,lr,it,x_orig):
    l=x.shape[1]
    m=np.zeros(l)
    for i in range(it):
        m=step(x,y,lr,m)
        c=m[l-1]    
        cost_n=cost(x_orig,y,m,c)
        print("Cost at it-:",cost_n,it)

    return m    
def predict(x,m):
    n=x.shape[1]
    c=m[n-1]
    c_prod=np.dot(x,m)
    return c_prod+c    
def run():
    data_test=np.loadtxt("test.csv",delimiter=",")
    x_f=data_test[:,:12]
    y_test=data_test[:,12]
    rows, cols = x_f.shape
    ones_column_test = np.ones((rows, 1), dtype=x_f.dtype)
    X_test = np.concatenate((x_f, ones_column_test), axis=1)
    iterations=20
    learning_rate=0.001
    m=gd(X_test,y_test,learning_rate,iterations,x_f)
    y_pred=predict(X_test,m)
    print("M",m)
    print("Predicted",y_pred)
    # print(X)
    # np.savetxt("predicted.csv",y_pred,delimiter=",",fmt='%.5f')
run()    
```
